```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Información general
tienda.info()

# Vista rápida de los primeros registros
tienda.head()

# Ver resumen estadístico de columnas numéricas
tienda.describe()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("Ingreso total por tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

### 2. Ventas por Categoría

En esta sección se analiza qué categorías de productos son las más vendidas en cada tienda. Para ello, se agrupan los datos por la columna `categoria` y se cuentan cuántas ventas hubo de cada tipo.

Esto permite identificar los productos más populares y entender el enfoque comercial de cada tienda.


```python
# Tienda 1
print("Tienda 1:")
print(tienda['Categoría del Producto'].value_counts())
print("\n")

# Tienda 2
print("Tienda 2:")
print(tienda2['Categoría del Producto'].value_counts())
print("\n")

# Tienda 3
print("Tienda 3:")
print(tienda3['Categoría del Producto'].value_counts())
print("\n")

# Tienda 4
print("Tienda 4:")
print(tienda4['Categoría del Producto'].value_counts())
```

### 3. Calificación Promedio de Clientes

En esta sección se calcula la calificación promedio que los clientes dieron a los productos de cada tienda. Esto permite evaluar el nivel de satisfacción de los compradores y detectar posibles problemas de calidad o atención.

El análisis se realiza a partir de la columna `evaluacion` de cada tienda.



```python
promedio_tienda1 = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

print("Evaluación promedio por tienda:")
print(f"Tienda 1: {promedio_tienda1:.2f}")
print(f"Tienda 2: {promedio_tienda2:.2f}")
print(f"Tienda 3: {promedio_tienda3:.2f}")
print(f"Tienda 4: {promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos
En esta sección se identifican los productos con mayor y menor cantidad de ventas en cada tienda. Analizar los productos más vendidos permite entender qué oferta tiene mayor aceptación, mientras que los menos vendidos pueden indicar productos poco atractivos o con problemas de comercialización.

Este análisis se basa en la columna `producto`, que contiene el nombre de cada artículo vendido.


```python
# Top 5 productos más vendidos - Tienda 1
print("Tienda 1 - Más vendidos:")
print(tienda['Producto'].value_counts().head(5))

# Bottom 5 productos menos vendidos - Tienda 1
print("\nTienda 1 - Menos vendidos:")
print(tienda['Producto'].value_counts().tail(5))
```

```python
# Top 5 productos más vendidos - Tienda 2
print("Tienda 2 - Más vendidos:")
print(tienda2['Producto'].value_counts().head(5))

# Bottom 5 productos menos vendidos - Tienda 2
print("\nTienda 2 - Menos vendidos:")
print(tienda2['Producto'].value_counts().tail(5))
```

```python
# Top 5 productos más vendidos - Tienda 3
print("Tienda 3 - Más vendidos:")
print(tienda3['Producto'].value_counts().head(5))

# Bottom 5 productos menos vendidos - Tienda 3
print("\nTienda 3 - Menos vendidos:")
print(tienda3['Producto'].value_counts().tail(5))
```

```python
# Top 5 productos más vendidos - Tienda 4
print("Tienda 4 - Más vendidos:")
print(tienda4['Producto'].value_counts().head(5))

# Bottom 5 productos menos vendidos - Tienda 4
print("\nTienda 4 - Menos vendidos:")
print(tienda4['Producto'].value_counts().tail(5))
```

# 5. Envío promedio por tienda

En esta sección se analiza el costo de envío promedio de cada tienda. Este indicador permite comprender cuánto gasta, en promedio, cada tienda para hacer llegar los productos a sus clientes. Tiendas con costos de envío muy altos podrían estar afectando su rentabilidad o satisfacción del cliente.


```python
envio_tienda1 = tienda['Costo de envío'].mean()
envio_tienda2 = tienda2['Costo de envío'].mean()
envio_tienda3 = tienda3['Costo de envío'].mean()
envio_tienda4 = tienda4['Costo de envío'].mean()

print("Costo de envío promedio por tienda:")
print(f"Tienda 1: ${envio_tienda1:.2f}")
print(f"Tienda 2: ${envio_tienda2:.2f}")
print(f"Tienda 3: ${envio_tienda3:.2f}")
print(f"Tienda 4: ${envio_tienda4:.2f}")
```

# Informe Final – Análisis de Tiendas Alura Store

## Objetivo

El Sr. Juan, propietario de la cadena de tiendas Alura Store, desea vender una de sus cuatro tiendas actuales para invertir en un nuevo negocio. Como analista de datos, mi tarea fue examinar el rendimiento de cada tienda para recomendar cuál debería vender.

Para ello, se analizaron cinco aspectos clave:
- Ingresos totales
- Categorías más populares
- Calificaciones promedio de clientes
- Productos más y menos vendidos
- Costo de envío promedio

## Análisis por Dimensiones

### 1. Ingresos Totales

| Tienda  | Ingreso Total     |
|---------|-------------------|
| Tienda 1 | $1,150,880,400.00 |
| Tienda 2 | $1,116,343,500.00 |
| Tienda 3 | $1,098,019,600.00 |
| **Tienda 4** | **$1,038,375,700.00** |

*Tienda 4 tiene el menor ingreso total, lo que podría indicar menor volumen de ventas o productos de menor valor.*


### 2. Ventas por Categoría

Las cuatro tiendas comparten patrones similares en sus categorías más vendidas, destacándose:
- **Muebles** y **Electrónicos** como las categorías dominantes.
- **Artículos para el hogar**, **Libros** e **Instrumentos musicales** tienen menor demanda en todas las tiendas.

La Tienda 4 presenta un volumen de ventas ligeramente inferior en la mayoría de las categorías clave, lo que reafirma un menor dinamismo comercial.



### 3. Evaluación Promedio de Clientes

| Tienda  | Evaluación Promedio |
|---------|----------------------|
| Tienda 1 | 3.98                |
| Tienda 2 | 4.04                |
| Tienda 3 | 4.05                |
| **Tienda 4** | **4.00**         |

*Aunque todas tienen buenas calificaciones, Tienda 4 tiene un desempeño ligeramente menor en satisfacción del cliente.*


### 4. Productos Más y Menos Vendidos

Tienda 4 destaca por tener productos con menos frecuencia de compra, incluso en categorías populares. Por ejemplo:
- Menos vendidos: Guitarra eléctrica (33), Armario (34), Refrigerador (38)
- Más vendidos: Cama box (62), Cubertería (59)

Las demás tiendas presentan productos con mayor rotación o equilibrio, indicando mejor aceptación de su catálogo.

---

### 5. Costo Promedio de Envío

| Tienda  | Costo Envío Promedio |
|---------|-----------------------|
| Tienda 1 | $26,018.61           |
| Tienda 2 | $25,216.24           |
| Tienda 3 | $24,805.68           |
| **Tienda 4** | **$23,459.46**   |

*Tienda 4 tiene el menor costo de envío promedio, lo cual es un punto a favor, pero no compensa sus debilidades en ventas y evaluación general.*



## Recomendación Final

Recomiendo al Sr. Juan vender la Tienda 4.

A pesar de tener el costo de envío más bajo, la Tienda 4:

- Tiene los ingresos más bajos.
- Registra un rendimiento inferior en ventas por categoría.
- Tiene calificaciones promedio más bajas.
- Exhibe menor rotación de productos.

Estas debilidades hacen que la Tienda 4 sea la opción menos rentable y con menor potencial de crecimiento a largo plazo, en comparación con las demás.



## Conclusión

La decisión está sustentada en datos reales y visualizaciones claras. Este análisis demuestra la importancia de la ciencia de datos aplicada a la toma de decisiones estratégicas.
